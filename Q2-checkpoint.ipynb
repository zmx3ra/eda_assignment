```python
## Question 4

descr = df['Price'].describe() # Describe age and save as an object
print(descr,'\n')

# Computing the Variance:
std = descr['std'] # Accessing the standard deviation from the description object
var = std**2
print('Variance: ', var,'\n')

# Computing the IQR:
thirdQ = descr['75%'] # Getting the 75 percentile
firstQ = descr['25%'] # Getting the 25 percentile
IQR = thirdQ - firstQ # Computing the IQR
print('IQR: ', IQR)
```

```python
## Question 3, Part 1

Df=pd.read_csv('./data/drilling_rigs.csv')

print(Df.shape)
Df.describe()

# there are 623 observations and 10 variables. Numeric variables are not being read in correctly by Pandas, so they need to be coerced. 
```

```python
Df.head(10)
```

```python
# Part 2

Df['time'] = pd.to_datetime(Df['Month'], format='mixed')
```

```python
Df['Active Well Service Rig Count (Number of Rigs)']=pd.to_numeric(Df['Active Well Service Rig Count (Number of Rigs)'], errors='coerce')
fig = sns.lineplot(data = Df, x = 'time', 
                   y = 'Active Well Service Rig Count (Number of Rigs)')
fig.set(title="Active Well Service Rig Count (Number of Rigs) over time", 
        xlabel='Date(Time)', ylabel='Number of Rigs')

#There is an overall decrease in number of Rigs over time, however if you zoom in it sometimes increasing vs. decreasing (going up and down).
```

```python
y=Df['Active Well Service Rig Count (Number of Rigs)'].to_numpy()
t=len(y)
rig_diff = y[1:]-y[:(t-1)]
fig = sns.lineplot(x=Df['time'][1:],y=rig_diff) # This is stationary
fig.set(title='Differences over Time: Number of Rigs First Difference')
plt.show()

#I definitely see a lot of variation from one year to the next, but over the whole graph we start and end around the same point.
```

```python
melted=Df.melt(id_vars='time', value_vars=['Crude Oil and Natural Gas Rotary Rigs in Operation, Onshore (Number of Rigs)','Crude Oil and Natural Gas Rotary Rigs in Operation, Offshore (Number of Rigs)'],var_name='Onshore or Offshore', value_name='Number of Rigs')
```

```python
melted.head(10)
```

```python
fig = sns.lineplot(data=melted,x='time',y='Number of Rigs',hue='Onshore or Offshore')
fig.set(title='Number of Rigs over time by Onshore or Offshore',xlabel='Year',ylabel='Number of Rigs')
sns.move_legend(fig, 'upper right', bbox_to_anchor=(1.7,1))
```
